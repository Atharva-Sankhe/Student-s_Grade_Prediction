```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('student-por.csv',sep=';')
df.head()
```

```python
df['school']=df['school'].str.split(";")
```

```python
df.shape
```

```python
df.iloc[0:1,1:2]=df['school'][0][1]
```

```python
for i in range(649):
    for j in range(1,33):
        df.iloc[i:i+1,j:j+1]=df['school'][i][j]
```

```python
df['school']=df['school'].str.get(0)
```

```python
df.to_csv('student_performance_cleaned.csv')
```

```python
df
```

```python
df['G1']=df['G1'].str.replace('"','')
df['G2']=df['G2'].str.replace('"','')
```

```python
for i in range(33):
    if df.iloc[0:1,i].str.isnumeric().all():
        df.iloc[:,i]=df.iloc[:,i].astype(int)
```

```python
df=pd.get_dummies(df,columns=['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic'],drop_first=True)
```

```python
x=df.drop(columns=['G3']).values
y=df['G3'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
mo=LinearRegression()
mo.fit(X_train,Y_train)
```

```python
Y_predict=mo.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,Y_predict)
```
